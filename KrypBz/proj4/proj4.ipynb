```python

def vypocitaj_prvocisla(strop):
    prvocisla = [3,5]
    for i in range(7,strop+1,2):
        kandidat = True
        for p in prvocisla:
            if i % p == 0:
                kandidat = False
                break
        if kandidat:
            prvocisla.append(i)
    return [2] + prvocisla
maximalna = 73*71
prvocisla = vypocitaj_prvocisla(maximalna)
print("Najdenych " + str(len(prvocisla)) + "prvocisel")
print(prvocisla)
```

```python
def vypocitaj_prvocisla2(hodnota):
    if hodnota % 2 == 0:
        return [2, hodnota / 2]
    prvocisla = [3,5]
    for i in range(7,hodnota+1,2):
        kandidat = True
        for p in prvocisla:
            if i % p == 0:
                kandidat = False
                break
        if kandidat:
            prvocisla.append(i)
            if hodnota % i == 0:
                prvocisla.append(hodnota / i)
            if len(prvocisla) % 100 == 0:
                print(prvocisla[-1:])
    return [2] + prvocisla

maximalna = 73*71
prvocisla = vypocitaj_prvocisla2(maximalna)
print("Najdenych " + str(len(prvocisla)) + "prvocisel")
print(prvocisla)
```

```python
def najdi_prvocisla(prvocisla, cislo):
    i = 0
    j = 0
    done = False
    for i in range(len(prvocisla) - 1):
        for j in range(i+1, len(prvocisla)):
            if prvocisla[i] * prvocisla[j] == cislo:
                done = True
                break
        if done:
            break
    return prvocisla[i], prvocisla[j]

a, b = najdi_prvocisla(prvocisla, maximalna)
print((a,b))
```

```python
def egcd(e,r):
    while(r!=0):
        e,r=r,e%r
    return e

#Euclid's Algorithm
def eugcd(e,r):
    for i in range(1,r):
        while(e!=0):
            a,b=r//e,r%e
            if(b!=0):
                print("%d = %d*(%d) + %d"%(r,a,e,b))
            r=e
            e=b

#Extended Euclidean Algorithm
def eea(a,b):
    if(a%b==0):
        return(b,0,1)
    else:
        gcd,s,t = eea(b,a%b)
        s = s-((a//b) * t)
        print("%d = %d*(%d) + (%d)*(%d)"%(gcd,a,t,s,b))
        return(gcd,t,s)

#Multiplicative Inverse
def mult_inv(e,r):
    gcd,s,_=eea(e,r)
    if(gcd!=1):
        return None
    else:
        if(s<0):
            print("s=%d. Since %d is less than 0, s = s(modr), i.e., s=%d."%(s,s,s%r))
        elif(s>0):
            print("s=%d."%(s))
        return s%r
```
